## yfinance Kütüphanesi ile Finans Analizi 

Yfinance kütüphanesi, Python programcılarının finans analizi yapmak için mükemmel bir kaynak sunar. Bu kütüphane, Yahoo Finans'tan finansal verileri kolayca çekmenizi, analiz etmenizi ve görselleştirmenizi sağlar.

```python
# Önce yfinance kütüphanesini Yükleyin
!pip install yfinance
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Apple Hisse Senetleri Verileri
apple = yf.Ticker("AAPL")

# Son haftanın Günlük Kapanış Verileri
daily_data = apple.history(period = "1wk", interval = "1d")

# Son 1 Yıl İçindeki Haftalık Kapanış Fiyatları için
# weekly_data = apple.history(period="1y", interval="1wk")

daily_data
```

```python
# Günlük Kapanış Fiyatlarına Ait Grafik
plt.figure(figsize=(12,6))
plt.plot(daily_data.index, daily_data["Close"], label="Apple Günlük Kapanış Fiyatı", color="blue")
plt.title("Apple Günlük Kapanış Fiyatları (Son 1 Hafta)")
plt.xlabel("Tarih")
plt.ylabel("Kapanış Fiyatı (USD)")
plt.legend()
plt.grid(True)
plt.show()
```
