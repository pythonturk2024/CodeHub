# How to create a Wordcloud in Python ?

A word cloud is a graphics method used to visually represent text data. Word clouds are a form of visual presentation that contains word labels that vary in size according to the frequency of words in a text document or dataset. The frequency of words is usually stated based on how often the word occurs in the text. That is, more frequently used words are displayed larger and more prominently, while less frequently used words are displayed smaller and less prominently.

Word clouds are commonly used to quickly visualize text data and highlight keywords or topics in texts. It is also useful for understanding which words are emphasized more or which topics stand out in a text.

The basic components of word clouds are:

**- Word Tags:** These are the main components of the word cloud. Words or terms in the text are represented by these tags.

**- Word Frequency:** The size of the labels changes depending on the frequency of the word used in the word clouds. That is, more frequently used words are shown with larger labels, while less frequently used words are shown with smaller labels.

**- Colors and Style:** It is possible to customize the colors and style of the word clouds. This can make the word cloud more attractive and meaningful.

**- Background Color:**

Sets the background color of the word cloud.

**- Color Map**

Determines the colors used in the word cloud. The color map is used to represent the different frequencies of words or tags in the word cloud.

```python
wordcloud = WordCloud(width=800, height=400, colormap='viridis')
```

**- Font Size**

Determines the size of words or tags. Larger font sizes make the word appear more prominent and larger in the word cloud.

```python
wordcloud = WordCloud(width=800, height=400, font_size=12)
```

**- Maximum Number of Words**

Determines the maximum number of words to be displayed in the word cloud.

```python
wordcloud = WordCloud(width=800, height=400, max_words=100)
```

**- Stopwords**

They are common and meaningless words that are ignored when mining text or creating word clouds. Example stop words can be conjunctions or common words such as "and", "or", "however".

```python
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(width=800, height=400, stopwords=stopwords)
```

**- Including Numbers**

Determines whether to display numbers in the word cloud.

```python
wordcloud = WordCloud(width=800, height=400, include_numbers=True)
```

## Here is a sample application with steps to create a word cloud using Python:

```python
# Import required libraries
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Create the data frame
data = {
    'text': [
        "Python programming language is a very useful and popular language. Python is used in data analysis, artificial intelligence, web development and many other fields. Learning Python is fun and useful.",
        "When data analysis is done with Python, libraries such as pandas, numpy and matplotlib are used. Pandas enables effective work on data frames.",
        "Python is a frequently used language in the field of artificial intelligence. It is preferred for machine learning and deep learning projects.",
        "For web development, Python is used with web frameworks such as Django and Flask. It is suitable for developing web applications."
    ]
}

df = pd.DataFrame(data)
```

```python
# Merge data
text = ' '.join(df['text'])

# Create the word cloud
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width=800, height=400, background_color='darkblue', colormap='coolwarm', max_words=100, stopwords=stopwords)
wordcloud.generate(text)

# Visualize the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
